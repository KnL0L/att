```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dias = np.arange(1,31)
vendas = np.random.randint(50,150, size=30)
```

```python
df_vendas = pd.DataFrame({'Dia': dias,'Vendas': vendas})
```

```python
df_vendas
```

```python
plt.figure(figsize=(12,10))
plt.plot(df_vendas['Dia'], df_vendas['Vendas'], marker='o', linestyle='-', color='b')

plt.title('Acompanhamneto de vendas diarias', fontsize=16)
plt.xlabel('Dias do Mês', fontsize=14)
plt.ylabel('Vendas Diarias', fontsize=14)
plt.grid(True)
plt.show()
```

```python
categoria = ['Eletronica', 'Roupas', 'Alimento', 'Moveis', 'Livros']
vendas_totais = [1200,850,950,501,750]
```

```python
df_vendas_categorias = pd.DataFrame({'Categoria': categoria, "Vendas": vendas_totais})
```

```python
plt.figure(figsize=(16,6))
plt.bar(df_vendas_categorias['Categoria'], df_vendas_categorias['Vendas'], color='skyblue')
plt.title('Vendas totais por Categoria de Produto - Mês Atual', fontsize=16)
plt.xlabel('Categoria de Produtos', fontsize=14)
plt.ylabel('vendas Totais', fontsize=14)
```
